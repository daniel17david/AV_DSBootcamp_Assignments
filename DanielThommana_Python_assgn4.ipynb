### Q 3 : Select the customers coming from state 'TN', identify the churning rate of such customers ?

```python
### Write your code here
churn_1 = data.loc[(data['Churn'] == 1) & (data['State'] == 'TN'), 'Churn'].value_counts()
churn_2 = data.loc[(data['Churn'] == 0) & (data['State'] == 'TN'), 'Churn'].value_counts()


churned=churn_1[1]
not_churned=churn_2[0]
overall=churn_1[1]+churn_2[0]
churning_rate=churned/overall
print("Overall:",overall)
print("Churned:",churned)
print("Not_Churnded:",not_churned)
print("churning_rate:",churning_rate)
```

### Q 7: Generate additional feature named F_X such that :-
* International plan is True then the value of F_X is 1
* If International plan is False and Account Length is more than the average Account length then F_X is 1
* If Number vmail messages is greater than 75th percentile and Customer Service calls is greater than 1 then F_X is 1
* In all other cases F_X is 0

```python
### Write your code here
def new_col(row):
  if row['International plan'] == True:
    return 1

  if row['International plan'] == False and row['Account length'] > data['Account length'].astype(str).str.len().mean():
    return 1

  if row['Number vmail messages'] > data['Number vmail messages'].quantile(0.75) and row['Customer service calls'] > 1:
    return 1

  return 0

data['F_X'] = data.apply(new_col, axis=1)
```
